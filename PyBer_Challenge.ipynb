### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "C:\\Users\\angel\\OneDrive\\Desktop\\Class_Folder\\Module_5\\PyBer_Analysis\\Resources\\city_data.csv"
ride_data_to_load = "C:\\Users\\angel\\OneDrive\\Desktop\\Class_Folder\\Module_5\\PyBer_Analysis\\Resources\\ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
# Dataframes for each city:
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
urban_cities_df.head()
```

```python
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
suburban_cities_df.head()
```

```python
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
rural_cities_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_ride_count
```

```python
# 2. Get the total drivers for each city type
total_driver_count = city_data_df.groupby(["type"]).sum()["driver_count"]
total_driver_count
```

```python
#  3. Get the total amount of fares for each city type
total_fare = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride = total_fare / total_ride_count
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = total_fare / total_driver_count
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
city_type_summary_df = pd.DataFrame({
        "Total Ride Count" : total_ride_count,
        "Total Driver Count" : total_driver_count,
        "Total Fare" : total_fare,
        "Average Fare Per Ride" : avg_fare_per_ride,
        "Average Fare Per Driver" : avg_fare_per_driver})
city_type_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
city_type_summary_df.index.name = None
```

```python
#  8. Format the columns.
city_type_summary_df["Total Ride Count"] = city_type_summary_df["Total Ride Count"].map("{:.0f}".format)
city_type_summary_df["Total Driver Count"] = city_type_summary_df["Total Driver Count"].map("{:.0f}".format)
city_type_summary_df["Total Fare"] = city_type_summary_df["Total Fare"].map("${:,.2f}".format)
city_type_summary_df["Average Fare Per Ride"] = city_type_summary_df["Average Fare Per Ride"].map("${:,.2f}".format)
city_type_summary_df["Average Fare Per Driver"] = city_type_summary_df["Average Fare Per Driver"].map("${:,.2f}".format)
city_type_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.


new_pyber_data_df = pyber_data_df.groupby(["type","date"]).sum()["fare"]
new_pyber_data_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
new_pyber_data_df = new_pyber_data_df.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
new_pyber_data_pivot = pyber_data_df.pivot(index ='date', columns = 'type', values = 'fare')
new_pyber_data_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
new_pyber_data_pivot = new_pyber_data_pivot.loc["2019-01-01":"2019-04-29"] 
new_pyber_data_pivot
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
new_pyber_data_df.index = pd.to_datetime(new_pyber_data_df.index)
new_pyber_data_df.index
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
new_pyber_data_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
new_pyber_data_df = new_pyber_data_df.resample('W').sum()
new_pyber_data_df.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

new_pyber_data_df.plot(figsize = (15,5))

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.

style.use('fivethirtyeight')
plt.title("Total Fare by City Type")
plt.ylabel("Fare ($USD)")
plt.xlabel("Months")

plt.savefig("C:\\Users\\angel\\OneDrive\\Desktop\\Class_Folder\\Module_5\\PyBer_Analysis\\analysis/PyBer_fare_summary.png", dpi= 300, bbox_inches='tight')
```
